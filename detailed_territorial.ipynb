```python
import matplotlib.pyplot as plt
from parse_tvk_data import *
import matplotlib.colors as mcolors
```

```python
regions={
    "західні області": west,
    "північні області": north,
    "східні області": east,
    "південні області": south,
    "центральні області": center,
    "київ":kyiv_city
}
for reg_name in regions.keys():
    raw=list(parse_election_tvk_data("vp2010","WP3355191.html?PT001F01=701",regions[reg_name]))
    prepared = prepare_data(
        raw,
        lambda row: int(row[2])/int(row[1]),
        {"янукович": 5, "тимошенко": 4, "проти всіх": 6}
    )

    fig=plt.figure()
    fig.set_size_inches(10 , 5, forward=True)
    fig.suptitle("Вибори 2010 року. "+reg_name)
    

    ax=fig.add_subplot(111)
    for name in prepared.keys():
        ax.plot(list(range(100)),prepared[name], label=name)
        pass
    ax.title.set_text("Другий тур")
    ax.set_xlabel("явка на дільницях")
    ax.set_ylabel("тисячі голосів на відсоток явки")
    ax.legend(loc=2)

    plt.draw()
    fig=plt.figure()
    fig.set_size_inches(10 , 5, forward=True)
    fig.suptitle("Другий тур виборів 2010 року. "+reg_name)
    

    julia = [int(row[4])/int(row[2])*100. if int(row[2])>0 else 0 for row in raw]
    vitya = [int(row[5])/int(row[2])*100. if int(row[2])>0 else 0 for row in raw]
    attendance=[int(row[2])/int(row[1])*100. if int(row[1])>0 else 0 for row in raw]
 
    ax=fig.add_subplot(121)
    ax.title.set_text("Голоси за Тимошенко")
    ax.set_xlabel("явка на дільницях")
    ax.set_ylabel("відсоток голосів")
    ax.hist2d(attendance, julia, bins=[50,50], norm=mcolors.PowerNorm(0.5))

    ax=fig.add_subplot(122)
    ax.title.set_text("Голоси за Януковича")
    ax.set_xlabel("явка на дільницях")
    ax.set_ylabel("відсоток голосів")
    ax.hist2d(attendance, vitya, bins=[50,50], norm=mcolors.PowerNorm(0.5))
    plt.draw()

```
